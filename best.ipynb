```
import pandas as pd
import numpy as np
import seaborn as sns
```

```
def get_date_features(df , column,name=""):
  df[column] = pd.to_datetime(df[column])
  df[name+'month'] = df[column].dt.month
  # df[name+'quarter'] = df[column].dt
  df[name+'day'] = df[column].dt.day
  return df
```

```
first_camp = pd.read_csv('Train/First_Health_Camp_Attended.csv')
first_camp = first_camp.rename(columns={"Donation":"First_camp_donation"})
first_camp.drop("Unnamed: 4",axis=1,inplace=True)
first_camp
```

```
second_camp = pd.read_csv('Train/Second_Health_Camp_Attended.csv')
second_camp = second_camp.rename(columns={"Health Score":"Health_Score"})

F_S_camp = pd.concat([first_camp,second_camp]).reset_index(drop=True)
F_S_camp['First_camp_donation'] = F_S_camp['First_camp_donation'].fillna(0)
F_S_camp
```

```
third_camp = pd.read_csv('Train/Third_Health_Camp_Attended.csv')
third_camp
```

```
patient_profile = pd.read_csv("Train/Patient_Profile.csv")
patient_profile.drop(['City_Type','Employer_Category'],axis=1,inplace=True)
patient_profile = get_date_features(patient_profile , "First_Interaction","first_interaction_")
# patient_profile.drop(['Income','Education_Score'],axis=1,inplace=True)
patient_profile.Age = patient_profile.Age.apply(lambda x:40 if x == "None" else x)
patient_profile.Education_Score = patient_profile.Education_Score.apply(lambda x:80 if x == "None" else x)
patient_profile.Income = patient_profile.Income.apply(lambda x:1 if x == "None" else x)

patient_profile
# sns.distplot(patient_profile[patient_profile.Income != "None"]['Income'].values)
```

```
camp_detail = pd.read_csv("Train/Health_Camp_Detail.csv")
camp_detail = get_date_features(camp_detail,"Camp_Start_Date",name="camp_start_")
camp_detail = get_date_features(camp_detail,"Camp_End_Date",name="camp_end_")
camp_detail['camp_duration_days'] = camp_detail.apply(lambda x:(x['Camp_End_Date'] - x['Camp_Start_Date']).total_seconds() / (60*60*24),axis=1)
camp_detail
```

```
train = pd.read_csv("Train/Train.csv")
train['is_train'] = True
test = pd.read_csv("test.csv")
test['is_train'] = False

merge = pd.concat([train,test]).reset_index(drop=True)
merge = merge.merge(patient_profile,on="Patient_ID")
merge = merge.merge(camp_detail , on= "Health_Camp_ID")

f_s_merge = merge[merge.Category1.isin(['First','Second'])].reset_index(drop=True).merge(F_S_camp,on=['Patient_ID','Health_Camp_ID'],how="left")
t_merge = merge[~merge.Category1.isin(['First','Second'])].reset_index(drop=True).merge(third_camp,on=['Patient_ID','Health_Camp_ID'],how="left")

f_s_merge = f_s_merge.fillna(0)
f_s_merge['outcome'] = f_s_merge.Health_Score.apply(lambda x:1 if x > 0 else 0)
f_s_merge.drop(['Health_Score'],axis=1,inplace=True)
f_s_merge

t_merge = t_merge.fillna(0)
t_merge['outcome'] = t_merge.Number_of_stall_visited.apply(lambda x:1 if x > 0 else 0)
t_merge.drop(['Number_of_stall_visited',"Last_Stall_Visited_Number"],axis=1 ,inplace=True)
# t_merge = t_merge.fillna(0) # dont change this.
t_merge

final = pd.concat([f_s_merge,t_merge])
final = final.fillna(0)
final.Category1 = pd.factorize(final.Category1)[0]
final.Category2 = pd.factorize(final.Category2)[0]
date_cols = ['Registration_Date','First_Interaction','Camp_Start_Date','Camp_End_Date']
for col in date_cols:
  final = get_date_features(final , col,col+"_")

final["days_left_for_camp_end_after_registration"] = final.apply(lambda x:(x['Camp_End_Date'] - x['Registration_Date']).total_seconds() / (60*60*24),axis=1)
final["days_camp_started_before_registration"] = final.apply(lambda x:(x['Registration_Date'] - x['Camp_Start_Date']).total_seconds() / (60*60*24),axis=1)
final['days_from_first_interaction'] = final.apply(lambda x:(x['Registration_Date'] - x['First_Interaction']).total_seconds() / (60*60*24),axis=1)
# final['days_for_camp_after_interaction'] = final.apply(lambda x:(x['Camp_Start_Date'] - x['First_Interaction']).total_seconds() / (60*60*24),axis=1)
# final['registration_before_or_after'] = final.apply(lambda x:1 if x['Registration_Date'] < x['Camp_Start_Date'] else 0 , axis=1)

final.drop(date_cols,inplace=True,axis=1)

train = final[final.is_train == True]
train.drop("is_train",axis=1,inplace=True)
train = train.reset_index(drop=True)
test = final[final.is_train == False]
test.drop("is_train",axis=1,inplace=True)
test = test.reset_index(drop=True)
```

```
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score ,train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score
from catboost import CatBoostClassifier
import warnings
warnings.filterwarnings("ignore")
```

```
X = train.drop(["outcome",'Patient_ID',"Health_Camp_ID"],axis=1)
Y = train[['outcome']]

def get_score(y_true,y_pred):
  return roc_auc_score(y_true,y_pred)

# models = [XGBClassifier(),LGBMClassifier(),LogisticRegression(),RandomForestClassifier()]

# for model in models:
#   # model.fit(X,Y)
#   score = cross_val_score(model,X,Y,scoring="roc_auc",cv=5,verbose=False)
#   print(f"{str(model)[:20]}",np.array(score).mean(),score)
# break
```

```
# x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,shuffle=True)
# model = LGBMClassifier()
from sklearn.model_selection import KFold

folds = KFold(n_splits=5,shuffle=True)
max_score = 0


for (train_idx,test_idx),i in zip(folds.split(X,Y),range(5)):
  print(f"working on Fold {i}")
  
  params = {
      "od_type":"Iter",
      "od_wait":50,
      "verbose":50
  }

  model = CatBoostClassifier(**params)
  model.fit(X.iloc[train_idx],Y.iloc[train_idx],
            cat_features = unique_cols , eval_set=(X.iloc[test_idx],Y.iloc[test_idx]))

  # model.fit(x_train,y_train)
  preds = model.predict_proba(X.iloc[test_idx])
  preds= [i[1] for i in preds]
  score = get_score(Y.iloc[test_idx],preds)
  if score > max_score:
    max_score = score
    best_idx = train_idx

  print("score :" ,score)
  print("---"*50)
```

```
# x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,shuffle=True)
# model = LGBMClassifier()
from sklearn.model_selection import KFold
params = {
    "od_type":"Iter",
    "od_wait":50,
    "verbose":50
}
val_idx = X[~X.index.isin(best_idx)].index.values.tolist()
val_idx
model = CatBoostClassifier(**params)
model.fit(X.iloc[best_idx],Y.iloc[best_idx],
          cat_features = unique_cols , eval_set=(X.iloc[val_idx],Y.iloc[val_idx]))

preds = model.predict_proba(X.iloc[val_idx])
preds= [i[1] for i in preds]
score = get_score(Y.iloc[val_idx],preds)

print("score :" ,score)
```
